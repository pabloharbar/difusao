```python
import matplotlib.pyplot as plt
import numpy as np

th = 5e-4
a = 0.02
b = 0.015
k = 20
Th = 40
Tinf = 20

h = np.linspace(0, 200, 1000)

n_eigenvalues = 100
eigenvectors = np.pi * (2 * np.array([i for i in range(n_eigenvalues)]) + 1) / (2 * a)

m_2 = 2 * h / (k * th)
Ci = [
    (Th - Tinf)
    * 4
    * (-1) ** (i + 1)
    / (np.cosh((ei**2 + m_2) ** 0.5 * b) * (2 * i - 1) * np.pi)
    for i, ei in enumerate(eigenvectors)
]
q_dot = (
    2
    * k
    * th
    * a
    * sum(
        [
            ci * (ei**2 + m_2) ** 0.5 * np.sinh((ei**2 + m_2) ** 0.5 * b)
            for ci, ei in zip(Ci, eigenvectors)
        ]
    )
)

plt.plot(q_dot, h)
plt.grid()
plt.show()
```
